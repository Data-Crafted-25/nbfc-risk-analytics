```python
import pandas as pd
import random

# Load existing agent file
agents = pd.read_csv("dim_agent.csv")

# Create dummy loan account IDs
loan_accounts = [f"LA{100000 + i}" for i in range(1, 501)]

rows = []

for _, agent in agents.iterrows():
    assigned_loans = random.sample(loan_accounts, random.randint(5, 15))
    
    for loan in assigned_loans:
        rows.append({
            "agent_id": agent["agent_id"],
            "agent_name": agent["agent_name"],
            "branch_id": agent["branch_id"],
            "role": agent["role"],
            "experience_years": agent["experience_years"],
            "agent_status": agent["agent_status"],
            "loan_account_id": loan
        })

# Create new dataframe
dim_agent_fixed = pd.DataFrame(rows)

# Save corrected CSV
dim_agent_fixed.to_csv("dim_agent_with_loan_account.csv", index=False)

print("✅ New dim_agent_with_loan_account.csv generated")
```

```python
import pandas as pd
import random

# Load existing loan account file (already generated in your pipeline)
fact_loan_account = pd.read_csv("fact_loan_account.csv")

loan_ids = fact_loan_account["loan_account_id"].tolist()

# Reload original agent dimension (without loan mapping)
dim_agent = pd.read_csv("dim_agent.csv")

rows = []

# Assign loans to agents
for loan_id in loan_ids:
    agent = dim_agent.sample(1).iloc[0]  # randomly assign an agent
    
    rows.append({
        "agent_id": agent["agent_id"],
        "agent_name": agent["agent_name"],
        "branch_id": agent["branch_id"],
        "role": agent["role"],
        "experience_years": agent["experience_years"],
        "agent_status": agent["agent_status"],
        "loan_account_id": loan_id
    })

# New dim_agent with loan mapping
dim_agent_mapped = pd.DataFrame(rows)

# Save corrected file
dim_agent_mapped.to_csv("dim_agent.csv", index=False)

print("✅ dim_agent.csv regenerated with correct loan_account_id mapping")
```
